```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data2 = pd.read_csv('power_mppt.csv')
data2
```

```python
data2 = data2.dropna()
data2
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data2['label'] = label_encoder.fit_transform(data2['label'])
print(data2['label'].unique())
```

```python
X = data2.drop(columns=['Time', 'label'])
Y = data2['label']
```

```python
X 
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold, ShuffleSplit
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, shuffle=True)
print('X_train ', X_train.shape, 'Y_train ', y_train.shape)
print('X_test ', X_test.shape, 'Y_test ', y_test.shape)
```

```python
import time
```

```python
scaler = StandardScaler()
X_sc_train = scaler.fit_transform(X_train)
X_sc_test = scaler.transform(X_test)

print('The number of samples in the training set {}'.format(len(X_sc_train)))
print('The number of samples in the test set {}'.format(len(X_sc_test)))
```

## DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(criterion='entropy',max_depth=10, min_samples_split=5, 
                                  min_samples_leaf=1, max_features=None,
                                  max_leaf_nodes=None, min_impurity_decrease=0.0)
t1 = time.time()
dt_model.fit(X_sc_train, y_train)
time_dt_model = time.time() - t1
print('Decision Tree time to training : {0:.5f}'.format(time_dt_model))
```

```python
y_pred_dt = dt_model.predict(X_sc_test)
print(y_pred_dt)
```

```python
cv = ShuffleSplit(n_splits=5, test_size=0.3)
score = cross_val_score(dt_model, X_sc_train, y_train, cv=cv, n_jobs=-1)
print(score)
print('{} Accuracy with a standard deviation of {}'.format(np.mean(score), np.std(score)))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, accuracy_score
from sklearn.metrics import roc_auc_score, roc_curve, auc
```

```python
dt_matrix = confusion_matrix(y_test, y_pred_dt)
normal = dt_matrix.astype('float') / dt_matrix.sum(axis=1)[:, np.newaxis]
disp = ConfusionMatrixDisplay(confusion_matrix=normal, display_labels=dt_model.classes_)
disp.plot()
plt.show()
```

```python
print(classification_report(y_test, y_pred_dt))
```

```python
for label in dt_model.classes_:
    y_test_label = y_test[y_test == label]
    y_pred_label = y_pred_dt[y_test == label]
    accuracy_label = accuracy_score(y_test_label, y_pred_label)
    print('Accuracy Fault {}: {:.3f}'.format(label, accuracy_label))
```

```python
y_scores = dt_model.predict_proba(X_sc_test)
auc_roc = roc_auc_score(y_test, y_scores, multi_class='ovr')
print("Mean AUC ROC:", auc_roc)

fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(len(label_encoder.classes_)):
    fpr[i], tpr[i], _ = roc_curve(np.array(pd.get_dummies(y_test))[:, i], y_scores[:,i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    
plt.figure(figsize=(5, 5))
for i in range(len(label_encoder.classes_)):
    plt.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %s' % (roc_auc[i], label_encoder.classes_[i]))

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating')
plt.legend(loc="lower right")
plt.show()
```

```python
train_errors = []
test_errors = []
depths = range(1, 11)# Thử nghiệm các giá trị cho max_depth từ 1 đến 10

for depth in depths:
    # Khởi tạo mô hình Decision Tree với độ sâu max_depth
    dt_model = DecisionTreeClassifier(criterion='entropy', max_depth=depth, min_samples_split=2, 
                                      min_samples_leaf=1, max_features=None,
                                      max_leaf_nodes=None, min_impurity_decrease=0.0)
    
    dt_model.fit(X_sc_train, y_train)# Huấn luyện mô hình trên tập huấn luyện
    train_error = 1 - dt_model.score(X_sc_train, y_train)# Đánh giá độ lỗi trên tập huấn luyện
    train_errors.append(train_error)
    test_error = 1 - dt_model.score(X_sc_test, y_test)# Đánh giá độ lỗi trên tập kiểm tra
    test_errors.append(test_error)

# Vẽ đồ thị
plt.figure(figsize=(5, 5))
plt.plot(depths, train_errors, label='Train error',color='red')
plt.plot(depths, test_errors, label='Test error',color='black')
plt.xlabel('Max Depth')
plt.ylabel('Error')
plt.title('Decision Tree Learning Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
#chọn siêu tham số cho Decision Tree
param_grid = {
                'criterion': ['gini', 'entropy'],
                'max_depth': [3, 5, 7, 10],
                'min_samples_split': [2, 5, 10],
                'min_samples_leaf': [1, 2, 4],
                'max_features': [None, 'sqrt', 'log2']}

grid_search = GridSearchCV(estimator=dt_model, param_grid=param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)
print('Best parameters found:')
print(grid_search.best_params_)
test_accuracy = grid_search.best_estimator_.score(X_test, y_test)
print('Decision Tree test accuracy with best parameters:', test_accuracy)
```

```python
from sklearn.model_selection import learning_curve

#Vẽ đồ thị học tập
def plot_learning_curve(estimator, X_sc_train, y_train):
    train_sizes, train_scores, test_scores = learning_curve(estimator, X_sc_train, y_train, train_sizes=np.linspace(0.1, 1.0, 10), cv=5)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    plt.figure(figsize=(5, 5))
    plt.plot(train_sizes, train_scores_mean, 'o-', color='r', label='Training score')
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color='r')
    plt.plot(train_sizes, test_scores_mean, 'o-', color='k', label='Cross-validation score')
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,test_scores_mean + test_scores_std, alpha=0.1, color='k')
    
    plt.xlabel('Training examples')
    plt.ylabel('Score')
    plt.title('Learning Curve')
    plt.legend()
    plt.show()
    
plot_learning_curve(dt_model, X_sc_train, y_train)
```

```python
# Vẽ biểu đồ Decision Tree
from sklearn.tree import plot_tree
feature_names = X.columns.tolist() #Xác định feature từ cột X
class_names = [str(label) for label in dt_model.classes_] # chuyển đổi nhãn thành list
plt.figure(figsize=(5, 5))
plot_tree(dt_model, filled=True, feature_names=feature_names, class_names=class_names)
plt.show()
```
