```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_feature = ['Time', 'Ipv', 'Vpv', 'Vdc', 'ia', 'ib', 'ic','va','vb','vc','Iabc','If','Vabc','Vf','label']
solar_power_iptt = pd.DataFrame(columns=column_feature)
solar_power_mppt = pd.DataFrame(columns=column_feature)

for num in range(0, 8):
    df1 = pd.read_csv('F{}L.csv'.format(num))
    df1['label'] = np.full((len(df1), 1), 'F{}L'.format(num))
    solar_power_iptt = pd.concat([solar_power_iptt, df1], ignore_index=True)
    
    df2 = pd.read_csv('F{}M.csv'.format(num))
    df2['label'] = np.full((len(df2), 1), 'F{}M'.format(num))
    solar_power_mppt = pd.concat([solar_power_mppt, df2], ignore_index=True)
```

## Intermediate Power Point Tracking

```python
ippt1 = solar_power_iptt.iloc[::10000,:]
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16,12))
axes = axes.flatten()

for i, col in enumerate(ippt1.columns[1:-1]):
    sns.scatterplot(data=ippt1, x='Time', y=col, hue='label', style='label', ec=None, palette='muted', edgecolor='White',ax=axes[i])
    axes[i].set_xlabel('Time')
    axes[i].set_ylabel(col)
    
for j in range(len(ippt1.columns[1:-1]), len(axes)):
    axes[j].axis('off')
    
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(data=solar_power_iptt, x='label', palette='husl')
plt.xlabel('Label')
plt.ylabel('Count')
plt.title('Distribution the number of Intermediate power point tracking')
plt.xticks(rotation=45)
plt.show()
```

```python
df1 = solar_power_iptt.drop(columns=['Time','label'])
plt.figure(figsize=(15,15))
ip = sns.heatmap(df1.corr(), annot=True, cmap='magma', fmt='.3', linewidths=0.8)
plt.title('Heatmap of correlation Ippt')
plt.show()
```

```python
label_ippt = solar_power_iptt['label'].value_counts(normalize=True) * 100
plt.figure(figsize=(5,5))
plt.pie(label_ippt, labels=label_ippt.index, autopct='%2.2f%%', startangle=140)
plt.title('Equal IPPT')
plt.show()
```
